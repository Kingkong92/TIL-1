```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax=plt.subplots()
tips=sns.load_dataset("tips")
tips
ax=sns.regplot(x="total_bill", y="tip", data=tips, fit_reg=False)
ax.set_xlabel("TB")
ax.set_ylabel("Tip")
ax.set_title("Total bill and Tip")
```

```python
joint=sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex")
joint.set_axis_labels(xlabel="TB", ylabel="Tip")
```

```python
#이차원 밀집도 그래프
kde, ax=plt.subplots()
ax=sns.kdeplot(data=tips['total_bill'],
           data2=tips['tip'], shade=True)
ax.set_title("Kernel Density Plot")
```

```python
ax=plt.subplots()
ax=sns.barplot(x="time", y="total_bill", data=tips)
```

```python
ax=plt.subplots()
sns.boxplot(x="total_bill", data=tips, orient="v")
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue="smoker", palette="Set3")
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue="smoker", palette="Set3")
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
ax=plt.subplots()
sns.violinplot(x="time", y="total_bill",data=tips)
```

```python
#관계그래프
sns.pairplot(tips)
```

```python
pg=sns.PairGrid(tips)
pg.map_upper(sns.regplot)
pg.map_lower(sns.kdeplot)
pg.map_diag(sns.distplot, rug=True)
```

```python
from sklearn.datasets import load_digits
digits=load_digits()
#digits
samples=[0,10,20,30,1,11,21,31]
```

```python
#d[7].shape
```

```python
d=[]
for i in range(8):
    d.append(digits.images[samples[i]])
```

```python
plt.figure(figsize=(8,2))
for i in range(8):
    plt.subplot(1,8,i+1)
    plt.imshow(d[i], cmap=plt.cm.bone_r)
    plt.title("image{}".format(i+1))
    plt.grid(False)
    plt.xticks([]);plt.yticks([])
```

```python
#2차원 이미지 => 64차원 데이터 => 1차원 벡터
```

```python
#스칼라와 벡터/행렬간 곱셈

#선형조합:벡터/행렬에 스칼라값을 곱한후 더하거나 뺀 것을
#벡터/행렬의 선형조합이라고 함.

#내적:두 벡터의 곱셈(dot)
#두 벡터의 차원이 같아야 함.
#행벡터 * 열벡터
#내적 결과는 스칼라
```

```python
x=np.array([[1],[2],[3]])
x
y=np.array([[4],[5],[6]])
y
#np.dot(x.T,y)
x.T@y
```

```python
#데이터벡터, 가중치벡터 => 데이터벡터의 가중치 합
```

```python
"""
벡터의 내적은 두 벡터의 유사도를 계산하는데 활용
유사도?두 벡터가 닮은 정도
코사인유사도?
"""
```

```python
digits=load_digits()
d1=digits.images[0]
d2=digits.images[10]
d3=digits.images[1]
d4=digits.images[11]
v1=d1.reshape(64,1)
v2=d2.reshape(64,1)
v3=d3.reshape(64,1)
v4=d4.reshape(64,1)
```

```python
plt.figure(figsize=(9,9)) #1당 72pix
import matplotlib.gridspec as gridspec
```

```python
gs=gridspec.GridSpec(1,8,height_ratios=[1],
                 width_ratios=[9,1,9,1,9,1,9,1])
```

```python
for i in range(4):
    plt.subplot(gs[2*i])
    plt.imshow(eval("d"+str(i+1)),cmap=plt.cm.bone_r)
    plt.subplot(gs[2*i+1])
    plt.imshow(eval("v"+str(i+1)),cmap=plt.cm.bone_r)
```

```python
v1 #0
(v1.T@v2)[0][0] #3064 :v1, v2벡터간 내적
(v1.T@v3)[0][0] #1866 :v1, v3벡터간 내적
(v1.T@v4)[0][0] #1883 :v1, v4벡터간 내적
```
