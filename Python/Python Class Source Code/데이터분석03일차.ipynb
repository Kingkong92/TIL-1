```python
dfleft = DataFrame({'KEY': ['K0', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']})


dfright = DataFrame({'KEY': ['K2', 'K3', 'K4', 'K5'],
    'C': ['C2', 'C3', 'C4', 'C5'],
    'D': ['D2', 'D3', 'D4', 'D5']})
```

```python
#dfleft, dfright를 KEY변수를 기준으로 merge
pd.merge(dfleft,dfright, how="inner")
#default : inner
```

```python
pd.merge(dfleft,dfright, how="left", on='KEY')
```

```python
pd.merge(dfleft,dfright, how="right", on='KEY')
```

```python
pd.merge(dfleft,dfright,
        how="left") #inner
```

```python
pd.merge(dfleft,dfright,
        how="right") #inner
```

```python
dfleft
```

```python
dfright
```

```python
pd.merge(dfleft,dfright, how='outer', on='KEY')#디폴트  how='inner'
```

```python
pd.merge(dfleft,dfright, how='outer', on='KEY', indicator=True)#디폴트  how='inner'
pd.merge(dfleft,dfright, how='outer', on='KEY', indicator='indicator_info')#디폴트  how='inner'
```

```python
#데이터프레임을 인덱스를 기준으로 병합
```

```python
dfleft = DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']}, 
index= ['K0', 'K1', 'K2', 'K3'])
dfright = DataFrame({
    'C': ['C2', 'C3', 'C4', 'C5'],
    'D': ['D2', 'D3', 'D4', 'D5']},
index= ['K2', 'K3', 'K4', 'K5'])
```

```python
#index 기준 병합 방법:join(), merge()
#index 기준 left join 수행

#merge()를 이용한 left join
pd.merge(dfleft, dfright, left_index=True, right_index=True)
pd.merge(dfleft, dfright, left_index=True, right_index=True, how='left')
pd.merge(dfleft, dfright, left_index=True, right_index=True, how='right')
```

```python
pd.merge(dfleft, dfright, left_index=True, right_index=True, how='right')

dfleft.join(dfright) #how=left
dfleft.join(dfright, how='right')
```

```python
dfleft = DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']}, 
index= ['K0', 'K1', 'K2', 'K3'])
dfright = DataFrame({
    'C': ['C2', 'C3', 'C4', 'C5'],
    'D': ['D2', 'D3', 'D4', 'D5']},
index= ['K2', 'K3', 'K4', 'K5'])
```

```python
#index기준 합치기, (merge함수 이용)
# K0
# K1
# K2
# K3

pd.merge(dfleft, dfright, left_index=True, right_index=True, how='left') #default : how='inner'
#join함수를 이용하여 동일한 결과 출력
dfleft.join(dfright,how='left')#default:how='left'
```

```python
#merge함수 이용, dfleft, dfright를 outer join
pd.merge(dfleft, dfright, left_index=True, right_index=True, how='outer') #default : how='inner'
#join함수 이용
dfleft.join(dfright, how="outer")
```

```python
#index와 key를 조합하여 merge
dfleft = DataFrame({
    'KEY': ['K0', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']})
dfright = DataFrame({
    'C': ['C2', 'C3', 'C4', 'C5'],
    'D': ['D2', 'D3', 'D4', 'D5']},
index= ['K2', 'K3', 'K4', 'K5'])
#dfright
pd.merge(dfleft, dfright, left_on='KEY', right_index=True, how="left")

dfleft.join(dfright, on='KEY', how='left')
```
