```python
dfleft = DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']}
                  , index=['K0','K1','K2','K3'])


dfright = DataFrame({
    'C': ['C2', 'C3', 'C4', 'C5'],
    'D': ['D2', 'D3', 'D4', 'D5']}
                   , index=['K2','K3','K4','K5'])
```

```python
dfright
```

```python
pd.merge(dfleft,dfright, left_index=True, right_index=True,how='left')
```

```python
dfleft.join(dfright)
```

```python
pd.merge(dfleft,dfright,left_index=True, right_index=True, how = 'outer')
```

```python
dfleft.join(dfright,how='outer')
```

```python
dfleft = DataFrame({'KEY':['K0','K1','K2','K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']}
                  )


dfright = DataFrame({
    'C': ['C2', 'C3', 'C4', 'C5'],
    'D': ['D2', 'D3', 'D4', 'D5']}
                   , index=['K2','K3','K4','K5'])
# col 은 on
# row 은 index
```

```python
pd.merge(dfleft,dfright, left_on='KEY',right_on=False,left_index=True,right_index=True,how='outer')
```

```python
pd.merge(dfleft,dfright, left_on='KEY',right_index=True,how='left')
```

```python
dfleft.join(dfright,on='KEY')
```

```python
dfleft.join(dfright)
```
