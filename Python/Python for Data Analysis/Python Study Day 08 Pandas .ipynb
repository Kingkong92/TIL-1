```python
#시계열 데이터 : DataFrame에서 index를 만들때
date_idx = pd.date_range("1/7/2020",periods=7,freq='D')
```

```python
df2 = pd.DataFrame({"c1" : [5,4,7,8,2,1,4]},index=date_idx)
df2
```
