```python
html="""
<html><body id="body">
<h1 id='title'>스크래핑</h1>
<p>웹 페이지 분석</p>
<p>원하는 부분 추출</p>
</body></html>
"""
from bs4 import BeautifulSoup

soup = BeautifulSoup(html,'html.parser')
```

```python
soup.html.body.h1
```

```python
print(soup.html.body.p)
print(soup.html.body.p)
## 형제 테크, level이 같은 테그로 감싸져있다.
## 두번째 p에 있는 원하는 부분을 추출하려면 next_sibling
p1 = soup.html.body.p
print(p1.next_sibling)  ## <p>웹 페이지 분석</p> 이 부분 뒤에 있는 \n 때문에 공백이 출력
print(p1.next_sibling.next_sibling)  ## 그래서 next를 2번 해줘야됨
# p1 -> \n -> 형제 태그
```

```python
# css 선택자 사용하기(크롬-> 개발자도구->객체선택->copy selector)사용
# soup.select_one(선택자) : 선택자로 지정된 요소 하나를 추출
# soup.select(선택자) : 선택자로 지정된 여러 개 요소를 추출
```

```python
html = """
<html>
    <body>
        <div id='myid'>
            <h1>2020년</h1>
            <ul class='day'>
                <li>Monday</li>
                <li>Tuesday</li>
                <li>Wednesday</li>
            </ul>
        </div>
    </body>
</html>
"""
```

```python
soup=BeautifulSoup(html,'html.parser')
```

```python
print(soup.select_one('h1'))
```

```python
print(soup.select_one("div#myid > h1").string)
## div 태그 내에 있는 h1을 출력
```

```python
print(soup.select_one("div#myid > ul.day > li") )
print(soup.select("div#myid > ul.day > li") )
# class 는 . 을 찍음 > 꺽새아님
# select_one 하면 하나만 출력됨
# select 로 하면 전부 출력 list 형태로
```

```python
st = soup.select("div#myid > ul.day > li") 
for a in st:
    print(a.string)
```
