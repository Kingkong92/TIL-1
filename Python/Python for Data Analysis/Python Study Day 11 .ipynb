```python
# 이항변수화 : 연속형변수를 // 기준이하 -> 0, 기준초가 -> 1
# 확률변수 x 가 이항분포를 따른다고 했을 떄
# 0또는 1값을 갖는 이항변수화가 필요
# 베르누이시행 : 성공 1 실패 0 두가지 경우만 존재
# 성공확률이 p인 베르누이시행을 n번 수행.
# 성공하는 횟수를 x라고하면,
# 확률변수 x는 모수 n과 p인 이항분포를 따른다.
```

```python
# 바이너라이저 : 연속형 변수를 이항변수화
# 원핫인코더 : 범주형 변수를 이항변수화
# 이항변수화 작업은 필수적인 작업니다.
'''
=====================================
원핫인코더 : 범주형 변수의 이항변수화 
=====================================

연령대 : 20대 - 0 , 30대 - 1, 40대 - 2
         012
         100 : 20대
         010 : 30대
         001 : 40대
         
성별 : 남 - 0 , 여 - 1
        10 남
        01 여

학점 : A:0, B:1, C:2, D:3, F:4  => 0과 1로만 되어있지 않음
       01234
       10000 : A
       01000 : B
       00100 : C
       00010 : D
       00001 : F
       길이를 5로 잡고 각 자리를 1로

연령대 & 성별 & 학점
20대 여성 B : 100 01 01000 -> 1000101000
30대 남성 D : 010 10 00010 -> 0101000010

'''
```

```python
df = pd.DataFrame({'C1': np.random.randn(20),

   'C2': ['a', 'a', 'a', 'a', 'a', 'a', 'a', 'a', 'a', 'a',

   'b', 'b', 'b', 'b', 'b', 'b', 'b', 'b', 'b', 'b']})

df
```

```python
'''
np.digitize함수, np.where 함수 : 연속형 변수 -> 이산화
이항변수화 : 0또는 1 값을 갖는 변수를 만드는 것
이산화 : 연속형 변수를 2개 이상의 범주를 갖는 변수로 변환
점수 : 75 -> A/B/C/D/F 범주(category) 나누는 것  : 이산화
'''
```

```python
np.random.seed(10)
np.random.randn(1)
```

```python
'''
df의 c1열 20개 data구성
10개 bin으로 균등하게 나눔
'''
df.C1.min()
df.C1.max()
bins = np.linspace(df.C1.min(), df.C1.max(), 10)
bins
```

```python
np.digitize(df.C1,bins) # df.C1의 값을 bins의 어느 구간에 해당되는지 확인
```

```python
df['C1_bin'] = np.digitize(df.C1,bins) 
df
```

```python
# 구간별로 요약 통계, 범주간 평균 차이
# 독립성검정, indexing 부여 등 다양하게 활용될 수 있다.
```

```python
df.groupby('C1_bin')['C1'].size() # C1_bin 열에 의해 그룹화했을 때
df.groupby('C1_bin')['C1'].mean() # C1 열의 크기, 그룹별 평균
df.groupby('C1_bin')['C1'].std()  # 데이터가 1건이면 표준편차가 없으므로 NaN
df.groupby('C1_bin')['C2'].value_counts()
df[df['C1_bin']==5] # 특정 그룹만
```

```python
# pd.get_dummies() : 가변수 생성 함수
pd.get_dummies(df['C1_bin']) # one hot encoding
```

```python
# 모델 = x1 * 가중치1 + x2 * 가중치2 + ... + x10 * 가중치10 + Bias
# 이런 형식으로 진행된다.
```

```python
pd.get_dummies(df['C1_bin'],prefix='C1')
```

```python
# np.where(조건, 참일때, 거짓일때)
df['h_l'] = np.where(df['C1'] >= df.C1.mean(),'high','low')
df # 두개의 범주로 나눈 것, 이산형화
```

```python
# 범주별로 요약 통계치를 구하고싶으면 groupby
df.groupby('h_l')['C1'].size() # h_l에 의해 그룹화하여 그에 대한 C1의 크기
df.groupby('h_l')['C1'].mean()
df.groupby('h_l')['C1'].std()
```

```python
# np.percentile 4분위 수를 구할 떄 쓰는 함수
```

```python
q1 = np.percentile(df.C1,25) 
q3 = np.percentile(df.C1,75) 
q1, q3
```

```python
# 연습문제 : C1 열 값 <= Q1 then    l
#            Q1  <  값  <= Q3 then  m
#            Q3 <   값  then        h
df['h_l'] = np.where(df['C1'] >= df.C1.mean(),'high','low')
df['h_m_l'] = np.where(df['C1'] > q1, np.where(df['C1'] > q3, 'h','m') ,'l' )
df
```

```python
# 실수 데이터 -> 카테고리화 / 범주화
# cut, qcut
# cut (실수 값의 경계 지정)
# qcut (똑같은 구간으로 나눔)
ages = [0,2,10,21,23,37,31,61,20,42,32,100]
bins = [1,15,25,35,60,99]
labels = ['미성년자','청년','중년','장년','노년']
cats = pd.cut(ages, bins, labels=labels)
cats
```

```python
cats.categories
```

```python
cats.codes
```

```python
df = pd.DataFrame(ages, columns=['ages'])
df['age_cat'] = pd.cut(ages,bins,labels = labels)
df
```
