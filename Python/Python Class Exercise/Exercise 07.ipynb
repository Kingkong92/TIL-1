```python
import re

f = open("C:/Users/Administrator/Desktop/BEXX0003.txt",'r',encoding = "utf-16")
file = f.read()
# pat = re.compile("(?P<author>[\<author\>])(?P<name>[가-핳]+)(?P<author2>[\<\/author\>])")
# m = pat.search(file)
# print(m.group('name'))

pat2 = re.findall("(?P<title>\<[head]+\>)(?P<sub_name>[가-핳][0-9]+[가-핳\s\W]+)(?P<title2>\<\/[head]+\>)",file)

for lines in pat2:
    print(lines[1])

pat5 = re.compile("양반|사람")

file2 = pat5.sub('사람',file)
    


pat4 = re.findall("\<\w\>[\"][가-핳_\s\W\d]+[\"]\<\/\w\>",file2)
for conv in pat4:
    print(conv.split('"')[1])

f2 = open("C:/Users/Administrator/Desktop/result.txt",'w')

for line in pat4:
    f2.write(line)
    
f2.close() 
f.close()
```
