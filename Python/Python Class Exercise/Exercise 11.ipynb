# Problem 01

파이썬 키워드에 대한 질문 총 갯수

```python
import urllib.request as req
from bs4 import BeautifulSoup
import re
import requests

url_nonum = "https://search.naver.com/search.naver?where=kin&kin_display=10&qt=&title=0&&answer=0&grade=0&choice=0&sec=0&nso=so%3Ar%2Ca%3Aall%2Cp%3Aall&query=%ED%8C%8C%EC%9D%B4%EC%8D%AC&c_id=&c_name=&sm=tab_pge&kin_start="

page = req.urlopen(url_nonum + '1')
doc = page.read()
soup = BeautifulSoup(doc,'html.parser')
numQ = soup.find("span", class_ = 'title_num').text
numQues = re.search('[\w\s\W]* (?P<num>[\s\d+\,]*[\d+])[건]',numQ)
num = int(numQues.group('num').replace(',',''))

print("=" * 50)
print('number of Questions : %d' % (num))
print("=" * 50)
```

# Problem 02

1페이지에 있는 질문 모두 추출

```python
def question_names(url):   # url을 넣으면 해당 페이지의 질문들을 출력하는 함수
    page = req.urlopen(url)   
    doc = page.read()
    soup = BeautifulSoup(doc,'html.parser')
    questions = soup.select("#main_pack > div.kinn.section._kinBase > ul dt.question a")
    for question in questions:   # 여러개의 질문들 차래대로 출력
        print(question.text)
```

```python
print("=" * 50)
print("First Page Queston list")
print("=" * 50)
question_names(url_nonum + '1')   # 첫 번째 페이지 결과 출력
print("=" * 50)
```

# Problem 03

1~10 페이지에 있는 질문 모두 추출

```python
print("=" * 50)
print("Questions from 1 page to 10 page")
for i in range(1,100,10):   # 각 페이지 끝자리가 1, 11, 21 ..
    print("=" * 50)          # 10단위로 올라가므로 차례대로 넣어줌
    print("page %s" %(str(int((i-1)/10)+1)))
    question_names(url_nonum + '%s' % (str(i)))  # 페이지 제목 출력 함수 사용
print("=" * 50)  
```

# Problem 04

1~10 페이지에 있는 질문들의 링크에 접속하여
해당 질문과 답변을 모두 추출

```python
def quesANS(url):   # 하나의 질문에 들어가 해당 질문과 답변을 출력하는 함수
    page = requests.get(url)    # 지식인 링크로 들어감  ## requests를 사용하는 것 주의 #주소 읽어서 들어가는데 encoding문제 때문에 
    soup = BeautifulSoup(page.text,'html.parser')
    soup_cont = soup.select('#content > div.question-content > div > div.c-heading._questionContentsArea.c-heading--default-old > div.c-heading__content')
    global back_Data     # 5번 문제를 풀기위한 것, 각 질문답변들을 모두 더하기 위함
    print("-" * 50,"Questions","-" * 50)
    for content in soup_cont:   # 질문 출력
        print(content.text)
        back_Data += content.text   # backData에 질문 저장
    
    print("-" * 50,"Answers","-" * 50)
    soup_ans = soup.select('#answer_1 > div._endContents.c-heading-answer__content > div._endContentsText.c-heading-answer__content-user')
    for ans in soup_ans:   # 답변 출력
        print(ans.text)
        back_Data += ans.text   # backData에 답변 저장
```

```python
def quesANSforPAGE(url):   # 페이지에 있는 여러개의 질문들의 링크를 가져오기 위함
    page = req.urlopen(url)  
    doc = page.read()
    soup = BeautifulSoup(doc,'html.parser')
    qs = soup.select("#main_pack > div.kinn.section._kinBase > ul dt.question a")
    urls = []
    for q in qs:
        urls.append(q.get('href'))  # url 저장함

    for url in urls:       # 각 질문의 url로 접속하여
        quesANS(url)        # 질문과 답변을 scraping한다
        print('\n\n\n')
```

```python
back_Data = ""
quesANSforPAGE(url_nonum + '1') # 첫 페이지의 모든 질문과 답변을 가져옴
```

```python
back_Data = ""
for i in range(1,100,10):   # 1~10페이지
    print("page %s" %(str(int((i-1)/10)+1)))  # 페이지 선택을 위한 것 1, 11, 21, 31
    quesANSforPAGE(url_nonum + '%s' % (str(i)))  # 해당 페이지의 모든 질문답변을 가져옴
```
